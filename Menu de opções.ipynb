```python
from time import sleep  

n1 = int(input('Digite o primeiro número: '))  
n2 = int(input('Digite o segundo número: '))   
opcao = 0
while opcao != 8:
    print('''[1]Soma
    [2]Subtração
    [3]Multiplicação
    [4]Divisão
    [5]Maior Número
    [6]Menor Número
    [7]Outro valores
    [8]Sair do Programa''')
    opcao = int(input(('Digite sua opção')))
    if opcao == 1:
        soma = n1 + n2
        print('A soma dos valores digitados é: {}'.format(soma))
    elif opcao == 2:
        subtracao = n1 - n2
        print('A subtração dos valores digitados é: {}'.format(subtracao))
    elif opcao == 3:
        multiplicacao = n1 * n2
        print('A Multiplicação dos valores digitados é: {}'.format(multiplicacao))
    elif opcao == 4:
        divisao = n1 / n2
        print('A Divisão dos valores digitados é: {:.1f}'.format(divisao))
    elif opcao == 5:
        if n1 > n2:
            maior = n1
        else:
            maior = n2
        print('O maior número entre os valores digitados é o {}'.format(maior))
    elif opcao == 6:
        if n1 < n2:
            menor = n1
        else:
            menor = n2
        print('O menor número entre os valores digitados é o {}'.format(menor))
    elif opcao == 7:
        print('Informe os númerows novamente: ')
        n1 = int(input("digite o primeiro valor: "))
        n2 = int(input('Digite o segundo valor: '))
    elif opcao == 8:
        print('FINALIZANDO O PROGRAMA...')
        sleep(3)
    else:
        print('Número digitado é inválido. Digite outro!')
print('Fim do programa. Volte Sempre')
```
